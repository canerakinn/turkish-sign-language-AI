```python
import cv2
import numpy as np
from PIL import Image, ImageTk
import tkinter as tk
from gtts import gTTS
import os
import pygame
import uuid
from collections import Counter
import torch
from matplotlib import pyplot as plt
import time

model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp9/weights/best.pt', force_reload=True)
model2 = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp15/weights/best.pt', force_reload=True)

last_20_frames = []
empty_results_count = 0
firstWord = ""
word = ""
modelBool=True


def process_frame(frame):
    global firstWord
    global word
    global last_20_frames
    global empty_results_count
    global modelBool
    upperCaseBool = False

    text_box.configure(state=tk.NORMAL)
    tempString = text_box.get("1.0", tk.END).strip()
    if tempString=="" or tempString[-1] =="." or tempString[-1]=="?":
        upperCaseBool = True
        
    text_box.configure(state=tk.DISABLED)
    
    #print(last_20_frames)
    
    if modelBool == True:
        results = model(frame)
    else:
        results = model2(frame)
        
    if len(results.pandas().xyxy[0]) == 0:
        empty_results_count += 1
    else:
        empty_results_count = 0

    if empty_results_count >= 10:
        last_20_frames.clear()
        empty_results_count = 0
    
    class_names = results.pandas().xyxy[0]["name"].tolist()
    last_20_frames.extend(class_names)
    
    if len(last_20_frames) > 20:
        last_20_frames = last_20_frames[-20:]
    
    counter = Counter(last_20_frames)
    
    if len(counter) > 0 and len(last_20_frames) == 20:
        most_common_class = counter.most_common(1)[0][0]
        print(most_common_class) 
        last_20_frames.clear()
        most_common_class= most_common_class.lower()
        
        
        
        
        
        if(most_common_class == "c2"):
            word="ç"
        elif(most_common_class == "g2"):
            word="ğ"
        elif(most_common_class == "i"):
            word="ı"
        elif(most_common_class == "i2"):
            word="i"
        elif(most_common_class == "n2"):
            word="n"
        elif(most_common_class == "o2"):
            word="ö"
        elif(most_common_class == "s2"):
            word="ş"
        elif(most_common_class == "u2"):
            word="ü"
        elif(most_common_class == "y2"):
            word="y"
        
        elif(most_common_class == "aksam"):
            word = "akşam"
        elif(most_common_class == "carpmak"):
            word = "çarpmak"
        elif(most_common_class == "hayir"):
            word = "hayır"
        elif(most_common_class == "hi"):
            word = "merhaba"
        elif(most_common_class == "kopek"):
            word = "köpek"
        elif(most_common_class == "kotu"):
            word = "kötü"
        elif(most_common_class == "nasil"):
            word = "nasıl"
        elif(most_common_class == "tesekkur"):
            word = "teşekkür ederim"
        elif(most_common_class == "yurumek"):
            word = "yürümek"
        elif(most_common_class == "nokta"):
            word="."
        else: word = most_common_class
        
        
        if upperCaseBool:
            word = word.capitalize()
            upperCaseBool=False
        
        if(word == "Iyi"):
            word = "İyi"
        
        #nasil+sen=nasılsın
        if word =="sen":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "nasıl" in current_text[-5:]:
                delete_last_word()
                word = "nasılsın ?" 
            if "Nasıl" in current_text[-5:]:
                delete_last_word()
                word = "Nasılsın ?" 
            
            text_box.configure(state=tk.DISABLED)

        #nasil+ben=nasılım
        if word =="ben":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "nasıl" in current_text[-5:]:
                delete_last_word()
                word = "nasılım ?" 
            if "Nasıl" in current_text[-5:]:
                delete_last_word()
                word = "Nasılım ?" 
            
            text_box.configure(state=tk.DISABLED)
            
        #spor+yapmak=spor yapacağım.
        if word =="yapmak":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "spor" in current_text[-4:]:
                delete_last_word()
                word = "spor yapacağım." 
            if "Spor" in current_text[-4:]:
                delete_last_word()
                word = "Spor yapacağım." 
            
            text_box.configure(state=tk.DISABLED)
            
        #köpek+yürümek=köpeğimi yürüyüşe çıkaracağım.
        if word =="yürümek":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "köpek" in current_text[-5:]:
                delete_last_word()
                word = "köpeğimi yürüyüşe çıkaracağım." 
            if "Köpek" in current_text[-5:]:
                delete_last_word()
                word = "Köpeğimi yürüyüşe çıkaracağım." 
            
            text_box.configure(state=tk.DISABLED)
            
        #araba+ben=arabam
        if word =="ben":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "araba" in current_text[-5:]:
                delete_last_word()
                word = "arabam" 
            if "Araba" in current_text[-5:]:
                delete_last_word()
                word = "Arabam" 
            
            text_box.configure(state=tk.DISABLED)
            
        #araba+sen=araban
        if word =="sen":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "araba" in current_text[-5:]:
                delete_last_word()
                word = "araban" 
            if "Araba" in current_text[-5:]:
                delete_last_word()
                word = "Araban" 
            
            text_box.configure(state=tk.DISABLED)

        #arabam+carpmak=arabamı çarptım.
        if word =="çarpmak":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "arabam" in current_text[-6:]:
                delete_last_word()
                word = "arabamı çarptım." 
            if "Arabam" in current_text[-6:]:
                delete_last_word()
                word = "Arabamı çarptım." 
            
            text_box.configure(state=tk.DISABLED)
        
        #araban+carpmak=arabanı çarptım.
        if word =="çarpmak":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "araban" in current_text[-6:]:
                delete_last_word()
                word = "arabanı çarptım." 
            if "Araban" in current_text[-6:]:
                delete_last_word()
                word = "Arabanı çarptım." 
            
            text_box.configure(state=tk.DISABLED)
            
        #iyi+ben=iyiyim
        if word =="ben":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "iyi" in current_text[-3:]:
                delete_last_word()
                word = "iyiyim" 
            if "İyi" in current_text[-3:]:
                delete_last_word()
                word = "İyiyim" 
            
            text_box.configure(state=tk.DISABLED)
        
        #kotu+ben=kötüyüm
        if word =="ben":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "kötü" in current_text[-4:]:
                delete_last_word()
                word = "kötüyüm" 
            if "Kötü" in current_text[-4:]:
                delete_last_word()
                word = "Kötüyüm" 
            
            text_box.configure(state=tk.DISABLED)
                
        #nerede+sen=neredesin ?
        if word =="sen":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "nerede" in current_text[-6:]:
                delete_last_word()
                word = "neredesin ?" 
            if "Nerede" in current_text[-6:]:
                delete_last_word()
                word = "Neredesin ?" 
            
            text_box.configure(state=tk.DISABLED)
       
        #nerede+ben=neredeyim ?
        if word =="ben":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "nerede" in current_text[-6:]:
                delete_last_word()
                word = "neredeyim ?" 
            if "Nerede" in current_text[-6:]:
                delete_last_word()
                word = "Neredeyim ?" 
            
            text_box.configure(state=tk.DISABLED)
            
        #çarpmak+ben=çarptım ?
        if word =="ben":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "nerede" in current_text[-6:]:
                delete_last_word()
                word = "neredesin ?" 
            if "Nerede" in current_text[-6:]:
                delete_last_word()
                word = "Neredesin ?" 
            
            text_box.configure(state=tk.DISABLED)
            
        #saat+nasıl=saat kaç ?
        if word =="nasıl":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "saat" in current_text[-4:]:
                delete_last_word()
                word = "saat kaçta ?" 
            if "Saat" in current_text[-4:]:
                delete_last_word()
                word = "Saat kaçta ?" 
            
            text_box.configure(state=tk.DISABLED)

        #ev+ben=evdeyim
        if word =="ben":
            text_box.configure(state=tk.NORMAL)
            current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
            print(current_text[-5])

            
            if "ev" in current_text[-2:]:
                delete_last_word()
                word = "evdeyim." 
            if "Ev" in current_text[-2:]:
                delete_last_word()
                word = "Evdeyim." 
            
            text_box.configure(state=tk.DISABLED)
     
        
        
        
        
        
        if modelBool:
            word = word+" "
        
        text_box.configure(state=tk.NORMAL)
        text_box.insert(tk.END, word)
        text_box.configure(state=tk.DISABLED)


        
    
    rendered_frame = np.squeeze(results.render())
    return rendered_frame



def change_mod():
    global modelBool
    modelBool = not modelBool

def del_letter():
    text_box.configure(state=tk.NORMAL)
    current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
    updated_text = current_text[:-1]  # Son harfi siler
    text_box.delete("1.0", tk.END)  # Mevcut içeriği siler
    text_box.insert("1.0", updated_text)  # Güncellenmiş içeriği ekler
    text_box.configure(state=tk.DISABLED)
    
def delete_last_word():
    text_box.configure(state=tk.NORMAL)
    current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
    words = current_text.split()  # İçeriği kelimelere böler
    if len(words) > 0:
        words.pop()  # Son kelimeyi listeden çıkarır
    updated_text = " ".join(words)  # Güncellenmiş içeriği yeniden birleştirir
    updated_text = updated_text+" "
    text_box.delete("1.0", tk.END)  # Mevcut içeriği siler
    text_box.insert("1.0", updated_text)  # Güncellenmiş içeriği ekler
    text_box.configure(state=tk.DISABLED)
    
def add_space():
    text_box.configure(state=tk.NORMAL)
    current_text = text_box.get("1.0", tk.END).strip()  # Mevcut içeriği alır ve boşlukları temizler
    updated_text = current_text + " "  # Sonuna bir boşluk ekler
    text_box.delete("1.0", tk.END)  # Mevcut içeriği siler
    text_box.insert("1.0", updated_text)  # Güncellenmiş içeriği ekler
    text_box.configure(state=tk.DISABLED)

def button_click_clear():
    text_box.configure(state=tk.NORMAL)
    text_box.delete("1.0", tk.END)  # Mevcut içeriği siler
    text_box.configure(state=tk.DISABLED)


def speak_text():
    text = text_box.get("1.0", tk.END).strip()
    unique_filename = str(uuid.uuid4()) + ".mp3"
    tts = gTTS(text=text, lang='tr')
    tts.save(unique_filename)

    # Sesin oynatılması
    pygame.mixer.init()
    pygame.mixer.music.load(unique_filename)
    pygame.mixer.music.play()


def button_click_exit():
    pygame.mixer.init()
    pygame.mixer.music.load("sound.mp3")
    pygame.mixer.music.play()

    # Oluşturulan MP3 dosyalarının silinmesi
    for filename in os.listdir("."):
        if filename.endswith(".mp3"):
            if filename != "sound.mp3":
                os.remove(filename)

    # Programın kapatılması
    video_capture.release()
    cv2.destroyAllWindows()
    root.destroy()



convert_to_gray = False

root = tk.Tk()

button_frame = tk.Frame(root)
button_frame.pack(side=tk.BOTTOM)

button_exit = tk.Button(button_frame, text="Mod Değiştir", command=change_mod)
button_exit.pack(side=tk.LEFT, padx=10, pady=10)

button_clear = tk.Button(button_frame, text="Temizle", command=button_click_clear)
button_clear.pack(side=tk.LEFT, padx=10, pady=10)

button_harf_sil = tk.Button(button_frame,text="Harf Sil", command = del_letter)
button_harf_sil.pack(side=tk.LEFT,padx=10,pady=10)

button_clear = tk.Button(button_frame, text="Kelime Sil", command=delete_last_word)
button_clear.pack(side=tk.LEFT, padx=10, pady=10)

button_yaz = tk.Button(button_frame, text="Boşluk", command=add_space)
button_yaz.pack(side=tk.LEFT, padx=10, pady=10)

button_speak = tk.Button(button_frame, text="Seslendir", command=speak_text)
button_speak.pack(side=tk.LEFT, padx=10, pady=10)

button_exit = tk.Button(button_frame, text="Çık", command=button_click_exit)
button_exit.pack(side=tk.LEFT, padx=10, pady=10)

text_box_frame = tk.Frame(root)
text_box_frame.pack(side=tk.BOTTOM, pady=10)

text_box = tk.Text(text_box_frame, height=5)
text_box.pack(fill=tk.BOTH, expand=True)
text_box.configure(state=tk.DISABLED)

label_video = tk.Label(root)
label_video.pack()


video_capture = cv2.VideoCapture(0)
while True:
    ret, frame = video_capture.read()
    if ret:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (640, 480))
        processed_frame = process_frame(frame)
        img = Image.fromarray(processed_frame)
        imgtk = ImageTk.PhotoImage(image=img)
        label_video.imgtk = imgtk
        label_video.configure(image=imgtk)

    root.update()

video_capture.release()
cv2.destroyAllWindows()
```
